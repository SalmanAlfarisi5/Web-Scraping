```python
import cloudscraper       
from bs4 import BeautifulSoup

url = 'https://www.carousell.sg/categories/computers-tech-213/?canChangeKeyword=true'
# Create a Cloudflare-aware session that mimics Chrome on Windows
scraper = cloudscraper.create_scraper(
    browser={'browser': 'chrome', 'platform': 'windows', 'desktop': True},
    delay=10,            # polite delay between solves
    interpreter='nodejs' # faster JS solver if Node is present
)

resp = scraper.get(url, timeout=30)
resp.raise_for_status()  # will raise if still blocked

soup = BeautifulSoup(resp.text, 'lxml')
```

```python
items = soup.find_all('div', class_='D_pp D_rO')
```

```python
item = soup.find('div', class_='D_pp D_rO')
print(item.prettify())
```

```python
for item in items:
    name = item.find('div', class_='D_qq').find('p', class_='D_jL D_jM D_jQ D_jT D_jW D_jY D_jU D_ki')
    price = item.find('div', class_='D_qN').p.text
    print(name)
    print(price)
```
