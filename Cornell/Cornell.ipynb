```python
import requests
from bs4 import BeautifulSoup
```

```python
request = requests.get('https://arxiv.org/search/?query=llm&searchtype=all&source=header')
print(request)
```

```python
# Simply Printing it
soup = BeautifulSoup(request.text)
papers = soup.find_all('li', class_="arxiv-result")
for index,paper in enumerate(papers): 
    id = paper.div.p.a.text
    pdf = paper.div.p.a['href']
    title = paper.find('p', class_='title is-5 mathjax').text.strip()

    authors_list = []
    authors = paper.find('p', class_='authors').find_all('a')
    for author in authors:
        authors_list.append(author.text)
    
    print(f"Paper title: {title} \n")
    print(f"Paper id: {id} \n")
    print(f"Paper link: {pdf} \n")
    print(f"Authors: {authors_list}")
    print('--------------------------------------------------------------------------')
```

```python
# Save the file
soup = BeautifulSoup(request.text)
papers = soup.find_all('li', class_="arxiv-result")
for index,paper in enumerate(papers): 
    id = paper.div.p.a.text
    pdf = paper.div.p.a['href']
    title = paper.find('p', class_='title is-5 mathjax').text.strip()

    authors_list = []
    authors = paper.find('p', class_='authors').find_all('a')
    for author in authors:
        authors_list.append(author.text)
    
    with open(f'papers/{id}.txt', 'w') as f:
        f.write(f"Paper title: {title} \n")    
        f.write(f"Paper id: {id} \n")
        f.write(f"Paper link: {pdf} \n")
        f.write(f"Authors: {authors_list}")
```
